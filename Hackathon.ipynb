```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('pick_time_warehouse_train - pick_time_warehouse_train.csv')
df.head(5)
```

```python
df.isnull().sum()
```

```python
df=df.fillna(method='ffill')
df.head(5)
```

```python
df.isnull().sum()
```

```python
li=[]
r=[]
for i in range(df.shape[0]):
    li.append(df['Start_Time_of_Picking'].iloc[i].split(':'))
for i in range(df.shape[0]):
    r.append((float(li[i][0])*60)+float(li[i][1]))
df2=pd.DataFrame(data=r)
df2
df['Start_Time_of_Picking']=df2
df['Start_Time_of_Picking']
```

```python
# df['Start_Time_of_Picking']=pd.to_datetime(df.Start_Time_of_Picking)
# df.head(5)
```

```python
x=df
x=x.drop(columns=['sl.no','Pick_Time','last_station_served_by_user'])
x.head(5)
```

```python
y=df.iloc[:,1]
y.head(5)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
x_train.head(5)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=18)
x_train=pca.fit_transform(x_train)
x_test=pca.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
