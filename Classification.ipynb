```python
import pandas as pd
```

```python
df=pd.read_csv('Breast+cancer+data.csv')
```

```python
df.isnull().sum()
```

```python
df
```

```python
x=df.iloc[:,2:32]
x
```

```python
y=df.iloc[:,1]
y
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=ss.fit_transform(x)
```

```python
x=pd.DataFrame(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

# SVC using SVM

```python
from sklearn.svm import SVC
classifier1=SVC(kernel='linear')
classifier1.fit(x_train,y_train)
y_pred=classifier1.predict(x_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score
print(accuracy_score(y_test,y_pred)*100)
print(classification_report(y_test,y_pred))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
classifier2=DecisionTreeClassifier(criterion='entropy',max_depth=5)
classifier2.fit(x_train,y_train)
y_pred1=classifier2.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred1)*100)
print(classification_report(y_test,y_pred1))
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
classifier3=LogisticRegression()
classifier3.fit(x_train,y_train)
y_pred2=classifier3.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred2)*100)
print(classification_report(y_test,y_pred2))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
classifier4=KNeighborsClassifier(n_neighbors=4)
classifier4.fit(x_train,y_train)
y_pred3=classifier4.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred3)*100)
print(classification_report(y_test,y_pred3))
```

# Kmeans Cluster

```python
from sklearn.cluster import KMeans
classifier5=KMeans(  n_clusters=2, init='random',n_init=10, max_iter=300, tol=1e-04, random_state=0)
classifier5.fit(x_train,y_train)
y_pred4=classifier5.predict(x_test)
```

```python
y_pred4=pd.DataFrame(y_pred4)
```

```python
from sklearn.preprocessing import LabelEncoder
ler=LabelEncoder()
y_tes=ler.fit_transform(y_test)
```

```python
y_tes=pd.DataFrame(y_tes)
```

```python
print(accuracy_score(y_tes,y_pred4)*100)
print(classification_report(y_tes,y_pred4))
```
